```python
import strawberryfields as sf
from strawberryfields.ops import *

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import cm
import math

mpl.rcParams['figure.facecolor'] = 'white'
```

```python
# teleportation
prog = sf.Program(3)

alpha = 1j
r = np.abs(alpha)
phi = np.angle(alpha)

with prog.context as q:
    Squeezed(r, phi) | q[0]
    Squeezed(-2) | q[1]
    Squeezed(2) | q[2]
    
    BSgate(np.pi/4, np.pi) | (q[1], q[2])
    BSgate(np.pi/4, np.pi) | (q[0], q[1])
    
    MeasureX | q[0]
    MeasureP | q[1]
    
    Xgate(np.sqrt(2) * q[0].par) | q[2]
    Zgate(-np.sqrt(2) * q[1].par) | q[2]

eng = sf.Engine('gaussian')#, backend_options={"cutoff_dim": 15})
result = eng.run(prog, shots=1, modes=None, compile_options={})

print(result.samples)

quad_axis = np.linspace(-5, 5, 1000)
W = result.state.wigner(mode=2, xvec=quad_axis * sf.hbar, pvec=quad_axis * sf.hbar)

color_range = np.max(W.real)
nrm = mpl.colors.Normalize(-color_range, color_range)
plt.axes().set_aspect("equal")
plt.contourf(quad_axis, quad_axis, W, 60, cmap=cm.RdBu, norm=nrm)
plt.xlabel("q", fontsize=15)
plt.ylabel("p", fontsize=15)
plt.title('teleported state')
plt.tight_layout()
plt.show()

# drawing the wigner function of the input state
prog0 = sf.Program(1)

with prog0.context as q:
    Squeezed(r, phi) | q[0]
    
eng0 = sf.Engine('gaussian')#, backend_options={"cutoff_dim": 15})
state0 = eng0.run(prog0, shots=1, modes=None, compile_options={}).state

W0 = state0.wigner(mode=0, xvec=quad_axis * sf.hbar, pvec=quad_axis * sf.hbar)

color_range = np.max(W0.real)
nrm = mpl.colors.Normalize(-color_range, color_range)
plt.axes().set_aspect("equal")
plt.contourf(quad_axis, quad_axis, W0, 60, cmap=cm.RdBu, norm=nrm)
plt.xlabel("q", fontsize=15)
plt.ylabel("p", fontsize=15)
plt.title('input state')
plt.tight_layout()
plt.show()

print('fidelity ' + str(result.state.fidelity((state0.means(), state0.cov()), mode=2)))
```

```python
# teleportation
prog = sf.Program(4)

alpha = 1j
r = np.abs(alpha)
phi = np.angle(alpha)

with prog.context as q:
    Squeezed(r, phi) | q[0]
    Squeezed(-2) | q[1]
    Squeezed(2) | q[2]
    Coherent(100) | q[3]
    
    BSgate(np.pi/4, np.pi) | (q[1], q[2])
    BSgate(np.pi/4, np.pi) | (q[0], q[1])
    
    MeasureX | q[0]
    MeasureP | q[1]
    
    mx = q[0].par
    my = q[1].par

    Dgate(10/np.sqrt(2)*sf.math.sqrt(mx**2+my**2)-100) | q[3]
    Rgate(-sf.math.atan(my/mx)) | q[3] # in practical, the range of the argument of this rotation gate must be from -\pi to \pi, not \pi/2 to \pi/2. 

    BSgate(sf.math.acos(np.sqrt(0.99)), np.pi) | (q[2], q[3])

eng = sf.Engine('gaussian')#, backend_options={"cutoff_dim": 15})
result = eng.run(prog, shots=1, modes=None, compile_options={})

print(result.samples)

quad_axis = np.linspace(-5, 5, 1000)
W = result.state.wigner(mode=2, xvec=quad_axis * sf.hbar, pvec=quad_axis * sf.hbar)

color_range = np.max(W.real)
nrm = mpl.colors.Normalize(-color_range, color_range)
plt.axes().set_aspect("equal")
plt.contourf(quad_axis, quad_axis, W, 60, cmap=cm.RdBu, norm=nrm)
plt.xlabel("q", fontsize=15)
plt.ylabel("p", fontsize=15)
plt.title('teleported state')
plt.tight_layout()
plt.show()

# drawing the wigner function of the input state
prog0 = sf.Program(1)

with prog0.context as q:
    Squeezed(r, phi) | q[0]
    
eng0 = sf.Engine('gaussian')#, backend_options={"cutoff_dim": 15})
state0 = eng0.run(prog0, shots=1, modes=None, compile_options={}).state

W0 = state0.wigner(mode=0, xvec=quad_axis * sf.hbar, pvec=quad_axis * sf.hbar)

color_range = np.max(W0.real)
nrm = mpl.colors.Normalize(-color_range, color_range)
plt.axes().set_aspect("equal")
plt.contourf(quad_axis, quad_axis, W0, 60, cmap=cm.RdBu, norm=nrm)
plt.xlabel("q", fontsize=15)
plt.ylabel("p", fontsize=15)
plt.title('input state')
plt.tight_layout()
plt.show()

print('fidelity ' + str(result.state.fidelity((state0.means(), state0.cov()), mode=2)))
```

The first one uses Xgate and Zgate to displace the output state. The second one uses the combination of a strong coherent light and a 99:1 beam splitter. The below one checks that both are almost the same displacement.

```python
# teleportation
prog = sf.Program(3)

with prog.context as q:
    Coherent(0) | q[0]
    Coherent(0) | q[1]
    Coherent(100) | q[2]
    
    mx = 1.66
    my = 2.9

    Xgate(np.sqrt(2) * mx) | q[0]
    Zgate(-np.sqrt(2) * my) | q[0]

    Dgate(5*sf.math.sqrt(2*(mx**2+my**2))-100) | q[2]
    Rgate(-sf.math.atan(my/mx)) | q[2]

    BSgate(sf.math.acos(np.sqrt(0.99)),np.pi) | (q[1], q[2])

eng = sf.Engine('gaussian')#, backend_options={"cutoff_dim": 15})
result = eng.run(prog, shots=1, modes=None, compile_options={})

quad_axis = np.linspace(-5, 5, 1000)
W0 = result.state.wigner(mode=0, xvec=quad_axis * sf.hbar, pvec=quad_axis * sf.hbar)

color_range = np.max(W0.real)
nrm = mpl.colors.Normalize(-color_range, color_range)
plt.axes().set_aspect("equal")
plt.contourf(quad_axis, quad_axis, W0, 60, cmap=cm.RdBu, norm=nrm)
plt.xlabel("q", fontsize=15)
plt.ylabel("p", fontsize=15)
plt.title('teleported state')
plt.tight_layout()
plt.show()

W1 = result.state.wigner(mode=1, xvec=quad_axis * sf.hbar, pvec=quad_axis * sf.hbar)

color_range = np.max(W1.real)
nrm = mpl.colors.Normalize(-color_range, color_range)
plt.axes().set_aspect("equal")
plt.contourf(quad_axis, quad_axis, W1, 60, cmap=cm.RdBu, norm=nrm)
plt.xlabel("q", fontsize=15)
plt.ylabel("p", fontsize=15)
plt.title('teleported state')
plt.tight_layout()
plt.show()
```

The success probability of the second method, which uses a strong coherent light and a BS, is 50% because arctan, whose range is from ${-\pi/2}$ to ${\pi/2}$, is used in the argument of the rotation gate, althogh the argument range is from ${-\pi}$ to ${\pi}$. If q[i].par can be used in the condition of is statement, which is practical, the success probability can be to 100%. 

```python
count = 0
for i in range(1000):
    prog = sf.Program(3)

    alpha = 0.5+0.8j
    r = np.abs(alpha)
    phi = np.angle(alpha)

    with prog.context as q:
        Squeezed(r, phi) | q[0]
        Squeezed(-2) | q[1]
        Squeezed(2) | q[2]
        
        BSgate(np.pi/4, np.pi) | (q[1], q[2])
        BSgate(np.pi/4, np.pi) | (q[0], q[1])
        
        MeasureX | q[0]
        MeasureP | q[1]

    eng = sf.Engine('gaussian')#, backend_options={"cutoff_dim": 15})
    result = eng.run(prog, shots=1, modes=None, compile_options={})

    if result.samples[0][0] >= 0:
        count += 1

print(count)
```
